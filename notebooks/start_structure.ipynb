```python
import pandas as pd
answer_df = pd.read_csv('../csv/test/test_answer.csv')
with open('../csv/test/test_CarBreakDown.csv', 'r') as file1:
    lines = file1.readlines()
with open('../csv/final_submission.csv', 'w') as file2:
    file2.write(f"id,breakdown_next_30_days\n")
    for i,line in enumerate(lines[1:],start=0):
        split_line = line.split(',')
        file2.write(f"{split_line[0]},{answer_df.iloc[i]['breakdown_next_30_days']}\n")
```
